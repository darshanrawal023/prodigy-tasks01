```python
# Importing the Pandas library
import pandas as pd

# Loading the dataset
olympic_data = pd.read_csv('/content/train.csv')
```

```python
#displaying the data
olympic_data.head(10)
```

```python
# Extract the 'Age' column
ages = olympic_data['Age'].dropna()  # Drop missing values
```

```python
 #Create a histogram to visualize the age distribution
plt.figure(figsize=(20, 8))
plt.hist(ages, bins=30, edgecolor='k', alpha=0.7)
plt.title('Age Distribution of Olympic Passengers')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.grid(axis='y', alpha=0.90)
plt.show()
```

```python
# Extract the 'Survived' column
survived_counts = olympic_data['Survived'].value_counts()

# Create a bar chart to visualize the survival status
plt.figure(figsize=(20, 8))
survived_counts.plot(kind='bar', color=['orange', 'blue'], alpha=0.5)
plt.title('Survival Status of Vortex Stats Passengers')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Number of Passengers')
plt.xticks(rotation=0)
plt.grid(axis='y', alpha=0.50)
plt.show()
```

```python
# Extract the 'Pclass' column
pclass_counts = olympic_data['Pclass'].value_counts()

# Create a bar chart to visualize the passenger class distribution
plt.figure(figsize=(15, 9))
pclass_counts.plot(kind='bar', color=['blue', 'orange', 'green'], alpha=0.5)
plt.title('Passenger Class Distribution on the Olympics')
plt.xlabel('Passenger Class')
plt.ylabel('Number of Passengers')
plt.xticks(rotation=0)
plt.grid(axis='y', alpha=0.50)
plt.show()
```

```python
# Extract the 'Embarked' column
embarked_counts = olympic_data['Embarked'].value_counts()

# Create a bar chart to visualize the embarkation ports
plt.figure(figsize=(7, 9))
embarked_counts.plot(kind='bar', color=['purple', 'pink', 'maroon'], alpha=0.5)  # Corrected 'maron' to 'maroon'
plt.title('Embarkation Ports of Titanic Passengers')
plt.xlabel('Embarkation Port')
plt.ylabel('Number of Passengers')
plt.xticks(rotation=0)
plt.grid(axis='y', alpha=0.50)
plt.show()
```
